```python
# Importing Libraries
from dotenv import load_dotenv
from anthropic import Anthropic
```

```python
# Load env variables
load_dotenv()
```

```python
# Create an API client
client = Anthropic()
model = "claude-sonnet-4-0"
```

```python
# Make a request
resp = client.messages.create(
    model=model,
    max_tokens=100,
    messages=[
        {
            "role": "user",
            "content": "What is cloud computing? Answer in a sentence."
        }
    ]
)
```
