### System prompts
System prompts are a powerful way to customize how Claude responds to user input. Instead of getting generic answers, you can shape Claude's tone, style, and approach to match your specific use case.

```python
# Install the dependencies
%pip install anthropic python-dotenv
```

```python
# Importing the required libraries
from anthropic import Anthropic
from dotenv import load_dotenv
```

```python
# Load env variables and create client
load_dotenv()
client = Anthropic()
```

```python
# Helper functions
def add_user_message(messages, text):
    messages.append({
        "role": "user",
        "content": text
    })

def add_assistant_message(messages, text):
    messages.append({
        "role": "assistant",
        "content": text
    })

def chat(messages):
    resp = client.messages.create(
        model="claude-sonnet-4-0",
        max_tokens=1000,
        messages=messages
    )
    return resp.content[0].text
```

```python
messages = []
add_user_message(messages, "How do I solve 5x + 2 = 3 for x?")

answer = chat(messages)
answer
```

```python
def chat_system_prompt(messages):
    system_prompt = """
    You are a patient math tutor.
    Do not directly answer a student's questions.
    Guide them to a solution step by step.
    """
    
    resp = client.messages.create(
        model="claude-sonnet-4-0",
        max_tokens=1000,
        messages=messages,
        system=system_prompt
    )
    return resp.content[0].text
```

```python
messages = []
add_user_message(messages, "How do I solve 5x + 2 = 3 for x?")

answer = chat_system_prompt(messages)
answer
```
