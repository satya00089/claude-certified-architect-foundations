```python
# importing libraries
from anthropic import Anthropic
from dotenv import load_dotenv
```

```python
# load env and create clients
load_dotenv()
client = Anthropic()
system_prompt = """
You are senior python developer
help user with python development, follow best practices.
avoid unnecessary information
"""
```

```python
messages = []

while True:
    user_text = input("> ")
    if user_text == "exit":
        break
    print(f"user: {user_text}")

    messages.append({
        "role": "user",
        "content": user_text
    })
    
    resp = client.messages.create(
        model="claude-sonnet-4-0",
        max_tokens=1000,
        messages=messages,
        system=system_prompt
    )
    
    print(f"assistant: {resp.content[0].text}")
    messages.append({
        "role": "assistant",
        "content": resp.content[0].text
    })
```
