```python
# Importing libraries

from anthropic import Anthropic
from dotenv import load_dotenv
from datetime import datetime
```

```python
# loading environment variables from .env file and create client
load_dotenv()
client = Anthropic()
```

```python
# get current date and time from claude.
messages = []
messages.append({
    "role": "user",
    "content": "what is the current date and time?"
})

resp = client.messages.create(
    model="claude-haiku-4-5",
    max_tokens=100,
    messages=messages
)
print(resp.content[0].text)
```

```python
# get current datetime in python

def get_current_datetime(date_format="%Y-%m-%d %H:%M:%S"):
    if not date_format:
        raise ValueError("date_format cannot be empty")

    return datetime.now().strftime(date_format)

print(get_current_datetime())
```

```python
from anthropic.types import ToolParam

get_current_datetime_schema = ToolParam({
    "name": "get_current_datetime",
    "description": "Get the current date and time in a specified format.",
    "input_schema": {
        "type": "object",
        "properties": {
            "date_format": {
                "type": "string",
                "description": "The format in which to return the date and time.",
                "default": "%Y-%m-%d %H:%M:%S"
            }
        },
        "required": [
            "date_format"
        ]
    },
})
```

```python
# get current date and time from claude.
messages = []
messages.append({
    "role": "user",
    "content": "what is the current date and time, formatted as %Y-%m-%d %H:%M?"
})

resp = client.messages.create(
    model="claude-haiku-4-5",
    max_tokens=100,
    tools=[get_current_datetime_schema],
    messages=messages
)
resp
```

```python
messages.append({
    "role": "assistant",
    "content": resp.content
})
```

```python
resp.content[0].input
```

```python
get_current_datetime(**resp.content[0].input)
```

```python
messages.append({
    "role": "user",
    "content": [{
        "type": "tool_result",
        "tool_use_id": resp.content[0].id,
        "content": get_current_datetime(**resp.content[0].input),
        "is_error": False
    }]
})
```
