## Technique

1. Add the user's instruction as a user message.
2. Add an assistant message that _looks like the start_ of the desired formatted content (e.g., ```json).
3. Call the API with a `stop_sequences` that prevents Claude from adding extra commentary (e.g., ```).

This causes Claude to continue the assistant message by producing only the requested content.


````python
from anthropic import Anthropic
from dotenv import load_dotenv
import json

load_dotenv()
client = Anthropic()

messages = []
messages.append(
    {"role": "user", "content": "Generate a short EventBridge rule as JSON"}
)
messages.append({"role": "assistant", "content": "```json"})
resp = client.messages.create(
    model="claude-sonnet-4-0", max_tokens=300, messages=messages, stop_sequences=["```"]
)
text = resp.content[0].text
parsed = json.loads(text.strip())
print(parsed)
````
